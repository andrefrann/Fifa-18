```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from math import *
import scipy.stats as stats
```

```python
data = pd.read_csv('CompleteDataset.csv')
data = data.dropna()
data = data[pd.notnull(data.Value) & pd.notnull(data.Overall) & pd.notnull(data.Potential)]
```

```python
Vm = []
Vk = []
V = []
for i in data.Value:
    if "M" in i:
        x = i.replace("€","")
        x = x.replace("M","")
        Vm.append(float(x))
            
    if "K" in i:
        x = i.replace("€","")
        x = x.replace("K","")        
        Vk.append(float(x))
        
for i in Vm:
    V.append(i*1000)
    
for i in Vk:
    V.append(i)
```

```python
Vlog = []
for i in V:
    Vlog.append(log(i))


Valor = pd.Series(V)
data["Valor"] = Valor
data = data[pd.notnull(data.Valor)]

Valor_log = pd.Series(Vlog)
data["Valorlog"] = Valor_log
data = data[pd.notnull(data["Valorlog"])]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

ax.scatter(data.Overall,data.Potential,data.Valor)
plt.title("Correlação overall x potencial x valor")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

ax.scatter(data.Overall,data.Potential,data.Valorlog)
plt.title("Correlação overall x potencial x valor(em log)")
plt.show()
```

```python
y = data.Valorlog
x = np.column_stack((data.Overall,data.Potential))  #Agrupa as variaveis preditorass

x = sm.add_constant(x, prepend=True) #Adiciona a coluna das constantes

res = sm.OLS(y,x).fit() #Cria e ajusta o modelo

res.summary()
```

```python
x1 = 0.2332
x2 = -0.0002
const = -9.0337


y = data.Overall*x1 + data.Potential*x2 + const

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot_wireframe(data.Overall,data.Potential,y)
plt.show()
```

```python
residuo = data.Valor - y
```

```python
plt.hist(residuo,bins = 100)
plt.title("Histograma de resíduos")
plt.xlim(0,50000)
plt.show()
```
